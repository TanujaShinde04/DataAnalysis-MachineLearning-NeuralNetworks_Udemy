## Matplotlib example workflow

```python
# import matplotlib get it ready for plotiing in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1.prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. setup plots
fig,ax= plt.subplots(figsize=(10,10))

# 3. plot the data
ax.plot(x,y)

# 4. customize plot
ax.set(title="simple plot",xlabel="x-axis",ylabel="yaxis")

# 5.save
fig.savefig("images/sample-plot.png")
```

```python
# make a plot from dictionary

nut_butter_prices= {"Almond butter":10,
                   "Peanut butter":8,
                   "Cashew butter":12}
nut_butter_prices
```

```python
fig,ax= plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Butter store prices",ylabel="price ($)");
```

```python
fig,ax= plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
```

```python
x= np.random.randn(1000)
fig,ax= plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# 1.

fig, ((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# 2.
fig,ax= plt.subplots(nrows=2,ncols=2,figsize=(10,5))

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
heart_disease= pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist(bins=20);
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50= heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
fig,ax= plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);

# ax.set_xlim([45,100])
```

```python
fig,ax= plt.subplots(figsize=(10,6))

scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);

# Customize plot
ax.set(title="Heart disease cholestrol level",xlabel="Age",ylabel="Cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(),title="target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
over_50.head()
```

```python
fig,(ax0,ax1)= plt.subplots(nrows=2,ncols=1,figsize=(10,10))


scatter= ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
ax0.set(title="heart disease and cholestrol level",xlabel="Age",ylabel="Cholestrol");
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(over_50["chol"].mean(),linestyle='--');

scatter= ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"]);
ax1.set(title="Heart disease Max heart rate",xlabel="Age",ylabel="Max heart rate");
ax1.legend(*scatter.legend_elements(),title="target");
ax1.axhline(over_50["thalach"].mean(),linestyle='--');
```
